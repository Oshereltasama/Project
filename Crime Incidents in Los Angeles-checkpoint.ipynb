## Analysis crime Incidents in Los Angeles:  

This dataset provides a detailed description of reported crime incidents in Los Angeles from 2020 to the present. The data, collected by the Los Angeles Police Department (LAPD), includes information on crime types, time of occurrence, location of the event, and demographic details of the victims. With over four years of data, this dataset is a valuable resource for researchers, data scientists, and analysts interested in examining crime trends, identifying patterns, and developing predictive models to enhance public safety.

In this analysis, we will examine criminal events in Los Angeles with a focus on trends and prominent phenomena. We will review the data to understand the extent of crime in the city, current trends, and the most problematic areas. Additionally, we will assess the impact of demographic and economic variables on crime and attempt to identify potential factors influencing changes in crime rates. The goal of this analysis is to provide a comprehensive and clear picture of the crime situation in the city and to propose ways to address the issue.

#### Distribution of Victim Ages:

```python
plt.figure(figsize=(15, 4))
plt.hist(df_cleaned['Vict Age'].dropna(), bins=50, edgecolor='black')
plt.title('Histogram of Vict Age (Cleaned Data)')
plt.xlabel('Vict Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Frequency Distribution of Victim Ages by Age Group:

```python
bins = range(0, 120, 5)  
labels = [f'{i}-{i+4}' for i in bins[:-1]] 
df['Age Group'] = pd.cut(df['Vict Age'], bins=bins, labels=labels, right=False)
age_group_counts = df['Age Group'].value_counts().sort_index()
plt.figure(figsize=(15, 4))
age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency Distribution of Victim Ages by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Focusing on at-risk populations - analysis of Crime Data Involving Victims Aged 0-4 

```python
df_filtered = df[df['Vict Age'].between(0, 4)]
min_age_filtered = df_filtered['Vict Age'].min()
max_age_filtered = df_filtered['Vict Age'].max()
summary_df_filtered = pd.DataFrame({
    'Age': ['Minimum ', 'Maximum '],
    'Value': [min_age_filtered, max_age_filtered]})
summary_df_filtered
```

#### Distribution by Location and Reporting District :

```python
plt.figure(figsize=(25, 8))
ax = sns.countplot(data=df_filtered, x='AREA NAME')
data_counts = df_filtered['AREA NAME'].value_counts()
top_areas = data_counts.nlargest(5).index
ax.grid(axis='y', linestyle='--', alpha=0.7)
for patch in ax.patches:
    area_label = patch.get_x() + patch.get_width() / 2
    area_name = ax.get_xticklabels()[int(area_label)].get_text()
    if area_name in top_areas:
        patch.set_color('red') 
    else:
        patch.set_color('lightblue')  
for patch in ax.patches:
    height = patch.get_height()
    ax.annotate(f'{int(height)}',
                xy=(patch.get_x() + patch.get_width() / 2, height),
                xytext=(0, 5),  
                textcoords='offset points',
                ha='center',
                va='bottom')
plt.title('Number of Crimes by Area (Age 0-4)')
plt.xlabel('Area Name')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=0)
plt.show()
```

##### We observe that the locations:
##### Pacific, Central, 77th Street, Newton, and N Hollywood
##### have the highest number of cases where the victim was aged 0-4

### High Prevalence Areas for Child Victims Aged 0-4 - Insights into Location-Based Crime Patterns:

#### There are several possible explanations for this phenomenon:
Population Concentration: Areas with a high number of young families or a young population may experience more incidents involving children aged 0-4.

Socioeconomic Conditions: Areas with lower socioeconomic status may have higher crime rates and violence, including crimes against young children, due to challenges such as poverty and poor economic conditions.

Access to Services: Areas with more daycare centers, medical facilities, and playgrounds might see higher incident rates involving young children due to more opportunities for such incidents.

Environmental Conditions: High density, elevated noise levels, and poor infrastructure in certain areas may contribute to higher rates of accidents and crime involving young children.

##### In light of the understanding that there are areas where relatively high numbers of cases involve victims aged 0-4, I recommend several steps to improve the situation. These recommendations are only an initial step, and solving the problem may require broader and more measured efforts.

Improving socioeconomic conditions in at-risk areas can reduce crime rates and enhance quality of life. Better employment opportunities and financial support reduce economic pressure, which in turn lowers the risk of crime and accidents. Additionally, developing support programs for young families, including parenting workshops and family counseling, helps improve parenting skills and reduce family-related issues, thus reducing crime-related risks and strengthening community ties. Moreover, investing in infrastructure such as parks, playgrounds, and transportation facilities contributes to safer environments for children, reduces accidents, and improves quality of life, which also helps mitigate crime-related risks.

## Analysis of Weapon Type:

```python
weapon_counts = df['Weapon Desc'].value_counts()
weapon_counts
```

##### I chose to focus on the top 10 most common weapon types to improve chart clarity and highlight the most important information. There are 79 different weapon types in total, and by concentrating on the top 10, I aim to streamline the analysis and make it more manageable and insightful:

```python
top_weapons = df['Weapon Desc'].value_counts().nlargest(10)
top_weapons = top_weapons.dropna()
total_incidents = top_weapons.sum()
top_4 = top_weapons.head(4)
colors = ['red' if weapon in top_4 else 'blue' for weapon in top_weapons.index]
plt.figure(figsize=(15, 8))
ax = top_weapons.plot(kind='barh', color=colors, legend=False)
plt.xlabel('Number of Incidents')
plt.ylabel('Weapon Type')
plt.title('Top 10 Most Common Weapon Types in Reports')
plt.gca().invert_yaxis()
for container in ax.containers:
    for patch in container:
        width = patch.get_width()
        percentage = (width / total_incidents) * 100
        ax.annotate(f'{percentage:.1f}%',
                    xy=(width, patch.get_y() + patch.get_height() / 2),
                    xytext=(5, 0),  
                    textcoords='offset points',
                    ha='left',
                    va='center')
plt.show()
```

```python
order = ['H', 'W', 'B', 'X', 'O', 'A']
count_data = df.groupby(['Vict Descent', 'Vict Sex']).size().reset_index(name='count')
threshold = 2000
filtered_data = count_data[count_data['count'] > threshold]
plt.figure(figsize=(15, 6))
ax = sns.countplot(data=df[df['Vict Descent'].isin(filtered_data['Vict Descent'])],
                   x='Vict Descent', hue='Vict Sex', palette='Set2', order=order)
ax.set_title('Distribution of Cases by Victim Descent and Gender', size=20)
ax.set_xlabel('Victim Descent')
ax.set_ylabel('Number of Cases')
plt.xticks(rotation=0)
for p in ax.patches:
    height = p.get_height()
    if height > threshold:
        ax.annotate(f'{height}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 5), 
                    textcoords='offset points')
plt.tight_layout()
plt.show()
```
